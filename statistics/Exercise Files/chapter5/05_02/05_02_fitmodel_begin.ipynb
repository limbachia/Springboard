Loading gapminder data for year 1985 (Live Aid!) and setting up plot as in chapter 3

```python
gapminder = pd.read_csv('gapminder.csv')
```

```python
gdata = gapminder.query('year == 1985')
gdata.head()
```

```python
size = 1e-6 * gdata.population

colors = gdata.region.map({'Africa': 'skyblue', 'Europe': 'gold', 'America': 'palegreen', 'Asia': 'coral'})

def plotdata():
    gdata.plot.scatter('age5_surviving','babies_per_woman',
                       c=colors,
                       s=size,
                       linewidths=0.5,
                       edgecolor='k',
                       alpha=0.5,
                       figsize=(8,6))
```

```python
plotdata()
```

### Fitting model using statsmodel library

1. **Only Intercept** model that estimates mean across all observations of the response variable (babies_per_woman)

```python
model = smf.ols(formula='babies_per_woman ~ 1',data=gdata)
grandmean = model.fit()
```

```python
def plotfit(fit):
    plotdata() # plots the original plot as shown above
    pp.scatter(gdata.age5_surviving,fit.predict(gdata),
               c=colors,s=30,
               linewidths=0.5,
               edgecolor='k',
               marker='D') # D = diamond
```

```python
plotfit(grandmean)
```

```python
grandmean.params
```

```python
gdata.babies_per_woman.mean()
```

The model indeed estimates the mean across all observations of the babies_per_woman.

2. Lets estimate means by region

```python
model = smf.ols(formula='babies_per_woman ~ 1 + region',data=gdata).fit()
```

```python
plotfit(model)
```

A mean has been estimated for each region.

```python
model.params
```

```python
gdata.groupby('region').babies_per_woman.mean()
```

The model considered one of the regions as the baseline and estimates means for other regions as an offset from the mean of the baseline region (Africa).

We can tell the model to treat all regions equally as:

```python
model = smf.ols(formula='babies_per_woman ~ -1 + region',data=gdata).fit()
```

```python
model.params
```

3. Add a quantitative variable to the model: age5_surviving

```python
surviving = smf.ols(formula='babies_per_woman ~ -1 + region + age5_surviving',data=gdata).fit()
```

```python
plotfit(surviving)
```

```python
surviving.params
```

It tells us that for every additional percentage point of child survival till age 5, the number of babies per woman decreses by 0.14.

The constant group terms are large because they theoretically represent the number of children per woman for a survival rate of 0% 

4. Different slop for every region using interaction term as oppose to the main term.

```python
surviving_byregion = smf.ols(formula='babies_per_woman ~ -1 + region + age5_surviving:region',data=gdata).fit()
```

```python
plotfit(surviving_byregion)
```

```python
surviving_byregion.params
```

Now there are four different slopes. India and China represented by large circles seem to be outliers. Perhaps. we can account for them by including by another main term, this time proportional to population. 

```python
surviving_byregion_population = smf.ols(formula='babies_per_woman ~ -1 + region + age5_surviving:region + population',data=gdata).fit()
```

```python
plotfit(surviving_byregion_population)
```

```python
surviving_byregion_population.params
```

Note: The diamonds don't lie simply along a straight line because of the population explanatory variable.
