# Visualizing regressions

### Regression
- lmplot()

### Residual
- residplot()

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house = pd.read_csv('../data/01-House-Price/train.csv')
```

## Regression
- lmplot()

```python
sns.set(font_scale=1.5)
a = sns.lmplot('TotalBsmtSF','SalePrice',data=house,hue='BldgType',palette='Set1',height=10,
           size=None)
```

Regression plot shows positve linear relationship between **t**otal **b**asement **a**rea (TBA, in sq.ft.) and house sale price (in USD) across all building types. However, relationship is strongest for townhouses with end unit (TwnhsE).
- Per unit increase in TBA leads to rise in the sale price acorss all building types, but highest in TwnhsE.  

## Residual
- residplot()

```python
g = sns.FacetGrid(house,col='BldgType',col_wrap=3,height=4,aspect=1,sharey=False,sharex=False)
g.map(sns.residplot,'TotalBsmtSF','SalePrice')
```

Regression of TBA on to sale price seems to capture reasonable amount of variance in the data for all building types except for bulding type 1Fam. For 1Fam, the variance in the residuals seems to increase with TBA.  

### Higher order regression

```python
fig, axes = plt.subplots(nrows=2,ncols=3,figsize=(15,10))
axes = axes.flatten()

for i,ax in enumerate(axes[:-1]):
    sel_house = house[house['BldgType']==house.BldgType.unique()[i]]
    ax.scatter(x = sel_house.TotalBsmtSF, y= sel_house.SalePrice,color = sns.color_palette()[i],alpha=0.5)
    
    sns.regplot('TotalBsmtSF','SalePrice',data=sel_house,scatter=None,
                color=sns.color_palette()[i],
                order=2,label='Second Order',ax=ax)
```

### Higher order residuals

```python
g = sns.FacetGrid(house,col='BldgType',col_wrap=3,height=4,aspect=1,sharey=False,sharex=False)
g.map(sns.residplot,'TotalBsmtSF','SalePrice',order=2)
```

# Visualizing univariate distributions
- Strip plots
- Swarm plots
- Boxplots 
- Violin plots

## stripplot()

Visualize distrubutions of sale prices of different building types

```python
plt.figure(figsize=(10,8))
sns.stripplot(x='BldgType',y='TotalBsmtSF',data=house,jitter=True,size = 10,alpha = 0.5)
```

## swarmplot()

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x='BldgType',y='SalePrice',data=house,size=4)
```

Let's also visualize another factor, type of masonary veneer used (BrCmn: common brick, BrkFace: face brick, CBlock: cinder block, and Stone)

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x='BldgType',y='SalePrice',data=house,size=4,hue='MasVnrType')
```

Changing orientation

```python
plt.figure(figsize=(10,8))
sns.swarmplot(y='BldgType',x='SalePrice',data=house,size=4,hue='MasVnrType',orient='h')
```

## violinplot()

```python
plt.figure(figsize=(10,8))
sns.stripplot(x='BldgType',y='SalePrice',data=house,size=4,jitter=True)
sns.violinplot(x='BldgType',y='SalePrice',data=house,inner=None,saturation=0,color='lightgray')
```

# Visualizing multivariate distributions
- Joint plots
- Pair plots
- Heat maps

## jointplot()
    - scatter
    - kde
    - regression
    - resudial

```python
from scipy.stats import pearsonr
p= sns.jointplot(x='TotalBsmtSF',y='SalePrice',data=house,height=7,stat_func=pearsonr)
print('Scatter')
```

```python
sns.jointplot(x='TotalBsmtSF',y='SalePrice',data=house,height=7,kind='kde')
print('kde')
```

```python
sns.jointplot(x='TotalBsmtSF',y='SalePrice',data=house,height=7,kind='regression')
print('regression')
```

```python
sns.jointplot(x='TotalBsmtSF',y='SalePrice',data=house,height=7,kind='residual')
print('residual')
```
