# Working with 2D arrays
- pcolor (pseudocolor plot)
- imshow

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
u = np.linspace(-2,2,3)
v = np.linspace(-1,1,5)
X,Y = np.meshgrid(u,v)

Z = X**2/25 + Y**2/4

print('Z:\n',Z)
plt.set_cmap('gray')
plt.pcolor(Z)
plt.show()
```

```python
# Generate two 1-D arrays: u, v
u = np.linspace(-2, 2, 41)
v = np.linspace(-1,1,21)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid(u,v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor()
plt.set_cmap('viridis')
plt.pcolor(Z)
plt.show()
```

## Visualizing bivariate functions
- contours
- contourf (filled contour plot)

```python
# Generate a default contour map of the array Z
plt.subplot(2,2,1)
plt.contour(X,Y,Z)

# Generate a contour map with 20 contours
plt.subplot(2,2,2)
plt.contour(X,Y,Z,20)

# Generate a default filled contour map of the array Z
plt.subplot(2,2,3)
plt.contourf(X,Y,Z)

# Generate a default filled contour map with 20 contours
plt.subplot(2,2,4)
plt.contourf(X,Y,Z,20)

# Improve the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
# Create a filled contour plot with a color map of 'viridis'
plt.subplot(2,2,1)
plt.contourf(X,Y,Z,20, cmap='viridis')
plt.colorbar()
plt.title('Viridis')

# Create a filled contour plot with a color map of 'gray'
plt.subplot(2,2,2)
plt.contourf(X,Y,Z,20, cmap='gray')
plt.colorbar()
plt.title('Gray')

# Create a filled contour plot with a color map of 'autumn'
plt.subplot(2,2,3)
plt.contourf(X,Y,Z,20, cmap='autumn')
plt.colorbar()
plt.title('Autumn')

# Create a filled contour plot with a color map of 'winter'
plt.subplot(2,2,4)
plt.contourf(X,Y,Z,20, cmap='winter')
plt.colorbar()
plt.title('Winter')

# Improve the spacing between subplots and display them
plt.tight_layout()
plt.show()
```

## Visualizing bivariate distributions

```python
import pandas as pd

house= pd.read_csv('../data/01-House-Price/train.csv')
num_house = house._get_numeric_data()
num_house.info()
```

```python
# Distribution of SalePrice
counts,bins, patches = plt.hist(house.SalePrice)
```

```python
# 2D histogram to see distributions of SalePrice and OverallQual

_=plt.hist2d(house.OverallQual,house.SalePrice)
plt.colorbar()
plt.xlabel('Overall Quality (1-10)')
plt.ylabel('Sale Price (USD)')
```

```python
# Distribution using hexagonal binning
plt.figure(figsize=(15,10))
plt.hexbin(house.OverallQual,house.SalePrice,gridsize=(20,10))
plt.colorbar()
plt.xlabel('Overall Quality (1-10)')
plt.ylabel('Sale Price (USD)')
```
