```python
import statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

Loading gapminder data for year 1985 (Live Aid!) and setting up plot as in chapter 3

```python
gapminder = pd.read_csv('gapminder.csv')
```

```python
gdata = gapminder.query('year == 1985')
```

```python
size = 1e-6 * gdata.population

colors = gdata.region.map({'Africa': 'skyblue', 'Europe': 'gold', 'America': 'palegreen', 'Asia': 'coral'})

def plotdata():
    gdata.plot.scatter('age5_surviving','babies_per_woman',
                      c=colors,s=size,linewidths=0.5,edgecolor='k',alpha=0.5)
```

Setting up model plot

```python
def plotfit(fit):
    plotdata()
    pp.scatter(gdata.age5_surviving,fit.predict(gdata),
               c=colors,s=30,linewidths=0.5,edgecolor='k',marker='D')
```

Three models from last video

```python
groupmeans = smf.ols(formula='babies_per_woman ~ -1 + region', data=gdata).fit()
```

```python
surviving = smf.ols(formula='babies_per_woman ~ -1 + region + age5_surviving', data=gdata).fit()
```

```python
surviving_byregion_population = smf.ols(
    formula='babies_per_woman ~ -1 + region + age5_surviving'
        '+ age5_surviving:region - age5_surviving + population',
    data=gdata).fit()
```

```python
surviving_byregion_population = smf.ols(
    formula='babies_per_woman ~ -1 + region + age5_surviving + age5_surviving:region + population',
    data=gdata).fit()
```

```python
plotfit(groupmeans)
```

```python
plotfit(surviving)
```

```python
plotfit(surviving_byregion_population)
```

How can we tell which of the above two models is better?

```python
pp.scatter(gdata.age5_surviving,groupmeans.resid)
pp.scatter(gdata.age5_surviving,surviving.resid)
pp.scatter(gdata.age5_surviving,surviving_byregion_population.resid)
```

#### MSE for each model

```python
for model in [groupmeans,surviving,surviving_byregion_population]:
    print(model.mse_resid)
```

The last model (surviving_by_region_population) has the smallest mse, hence it provides the best fit

#### Rsquared for each model

```python
for model in [groupmeans,surviving,surviving_byregion_population]:
    print(model.rsquared)
```

#### F statistic
Explanatory power of fit parameters compared to "random" fit vectors.

An F = 1 means that the parameters contributes the same as the random variables.  
A large F means the parameters explain much more than the randome variables

```python
for model in [groupmeans,surviving,surviving_byregion_population]:
    print(model.fvalue)
```

It looks like the surviving model hits a sweet spot.

```python
print(surviving.summary())
```

```python
sm.stats.anova_lm(groupmeans)
```

**df**:
- 1st row: shows number of degrees of freedom, the number of parameters (4 regions)
- 2nd row: shows number of degrees of freedom for residuals; number of observations minus number of parameters

**sum_sq & mean_sq**:
- Total and average explained residual variance

**F**:
- F statistics

**PR(>F)**:
- P value for the null model with same number of parameters, but where all the terms are random.

Clearly the groupmeans region explain a lot of the variance of the data, with very high F statistic.

```python
sm.stats.anova_lm(surviving)
```

The new term age_surviving explains a lot less of the variance, but the F statistics is still very high.

```python
sm.stats.anova_lm(surviving_byregion_population)
```

The population term explains far less variance, also it F statistics is also very low.
