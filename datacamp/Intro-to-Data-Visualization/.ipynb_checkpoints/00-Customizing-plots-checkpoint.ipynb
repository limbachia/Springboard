# Potting multiple graphs
- Plotting many graphs on common axes
- Creating axes within a figure
- Creating subplots within a figure

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Loading different datasets to perform various visualization for this exercise**

```python
house = pd.read_csv('../data/01-House-Price/train.csv')
```

```python
num_house = house._get_numeric_data()
print(num_house.info())
```

**Lets plot SalePrice vs. TotalBsmtSF**

```python
import numpy as np
import matplotlib as mpl

cmap = plt.cm.jet # define a colormap
# extract all colors from the jet map
cmaplist = [cmap(i) for i in range(cmap.N)]

# create the new map
cmap = mpl.colors.LinearSegmentedColormap.from_list(
    'Custom cmap', cmaplist, cmap.N)

# define the bins and normalize. Here we define bins using the min and max OverallOual
bounds = np.linspace(house.OverallQual.min(), house.OverallQual.max(), house.OverallQual.nunique())
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

plt.figure(figsize = (10,8))
a = plt.scatter(house['TotalBsmtSF'], house['SalePrice'],s=75, alpha = 0.5,c=house['OverallQual'],cmap=cmap,norm=norm)
plt.title('House Sale Price vs. Total Basement Area',fontdict=dict(size=18))
plt.tick_params(axis='both',labelsize =16)
plt.xlabel('Area (feet$^{2}$)',fontsize = 18)

actual_scale = range(0,int(1e6),int(1e5))
labeled_scale = [el/(1e3) for el in list(range(0,int(1e6),int(1e5)))]
plt.yticks(actual_scale,labels=labeled_scale)
plt.ylabel('Thousands (USD)',fontsize = 18)

# Annotating the house with house that has the maximal OverallQual and TotalBsmtSF, but mininal SalePrice
# Check the next cell to know how this house was found
plt.annotate('Best deal',xy=(6110,160000),
             xytext=(4500,300000),
             arrowprops={'color':'red'},
             fontsize=18)


## add colorbar to the side of the figure
cbar_axes = plt.axes([0.95,0.1,0.05,.8])
b = mpl.colorbar.ColorbarBase(cbar_axes,
                          cmap=cmap,
                          norm=norm,
                          spacing='proportional',
                          ticks=bounds,
                          boundaries=bounds)
cbar_axes.set_ylabel('Overall Quality',fontsize=18)


# Creating inset to zoom in on the beat deal house.
plt.axes([0.55,0.5,0.3,0.3])
a = plt.scatter(house['TotalBsmtSF'], house['SalePrice'],s=75, alpha = 0.25,c=house['OverallQual'],cmap=cmap,norm=norm)

plt.yticks(actual_scale[:7],labels=labeled_scale[:7])

_=plt.axis([4000, 7000, 0, 6e5])
```

**Find house with higest OverallQual and TotalBsmtSF, but lowest price, to be annoted in the plot above**

```python
house[['SalePrice','TotalBsmtSF','OverallQual']].sort_values(by=['OverallQual','TotalBsmtSF','SalePrice'],
                                                             ascending=[False,False,True]).iloc[0,:]
```

**Lets create the above plot using the ggplot style sheet**

```python
plt.style.use('ggplot')

plt.figure(figsize = (10,8))
a = plt.scatter(house['TotalBsmtSF'], house['SalePrice'],s=75, alpha = 0.25,c=house['OverallQual'],cmap=cmap,norm=norm)
plt.title('House Sale Price vs. Total Basement Area',fontdict=dict(size=18))
plt.tick_params(axis='both',labelsize =16)
plt.xlabel('Area (feet$^{2}$)',fontsize = 18)

actual_scale = range(0,int(1e6),int(1e5))
labeled_scale = [el/(1e3) for el in list(range(0,int(1e6),int(1e5)))]
plt.yticks(actual_scale,labels=labeled_scale)
plt.ylabel('Thousands (USD)',fontsize = 18)

# Annotating the house with house that has the maximal OverallQual and TotalBsmtSF, but mininal SalePrice
# Check the next cell to know how this house was found
plt.annotate('Best deal',xy=(6110,160000),
             xytext=(4500,300000),
             arrowprops={'facecolor':'red'},
             fontsize=18)


## add colorbar to the side of the figure
cbar_axes = plt.axes([0.95,0.1,0.05,.8])
b = mpl.colorbar.ColorbarBase(cbar_axes,
                          cmap=cmap,
                          norm=norm,
                          spacing='proportional',
                          ticks=bounds,
                          boundaries=bounds)
cbar_axes.set_ylabel('Overall Quality',fontsize=18)


# Creating inset to zoom in on the beat deal house.
plt.axes([0.55,0.5,0.3,0.3])
a = plt.scatter(house['TotalBsmtSF'], house['SalePrice'],s=75, alpha = 0.25,c=house['OverallQual'],cmap=cmap,norm=norm)

plt.yticks(actual_scale[:7],labels=labeled_scale[:7])

_=plt.axis([4000, 7000, 0, 6e5])
plt.style.use('default') # set style back to deafult
```
