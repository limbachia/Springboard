# Working with 2D arrays
- pcolor (pseudocolor plot)
- imshow

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
u = np.linspace(-2,2,3)
v = np.linspace(-1,1,5)
X,Y = np.meshgrid(u,v)

Z = X**2/25 + Y**2/4

print('Z:\n',Z)
plt.set_cmap('gray')
plt.pcolor(Z)
plt.show()
```

```python
# Generate two 1-D arrays: u, v
u = np.linspace(-2, 2, 41)
v = np.linspace(-1,1,21)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid(u,v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor()
plt.set_cmap('viridis')
plt.pcolor(Z)
plt.show()
```

## Visualizing bivariate functions
- contours
- contourf (filled contour plot)

```python
# Generate a default contour map of the array Z
plt.subplot(2,2,1)
plt.contour(X,Y,Z)

# Generate a contour map with 20 contours
plt.subplot(2,2,2)
plt.contour(X,Y,Z,20)

# Generate a default filled contour map of the array Z
plt.subplot(2,2,3)
plt.contourf(X,Y,Z)

# Generate a default filled contour map with 20 contours
plt.subplot(2,2,4)
plt.contourf(X,Y,Z,20)

# Improve the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
# Create a filled contour plot with a color map of 'viridis'
plt.subplot(2,2,1)
plt.contourf(X,Y,Z,20, cmap='viridis')
plt.colorbar()
plt.title('Viridis')

# Create a filled contour plot with a color map of 'gray'
plt.subplot(2,2,2)
plt.contourf(X,Y,Z,20, cmap='gray')
plt.colorbar()
plt.title('Gray')

# Create a filled contour plot with a color map of 'autumn'
plt.subplot(2,2,3)
plt.contourf(X,Y,Z,20, cmap='autumn')
plt.colorbar()
plt.title('Autumn')

# Create a filled contour plot with a color map of 'winter'
plt.subplot(2,2,4)
plt.contourf(X,Y,Z,20, cmap='winter')
plt.colorbar()
plt.title('Winter')

# Improve the spacing between subplots and display them
plt.tight_layout()
plt.show()
```

## Working with images

```python
img = plt.imread('../data/sunflower.jpg')

print(img.shape)

plt.imshow(img)
plt.axis('off')
```

**covert into grayscale**

```python
# method 1
collapsed = img.mean(axis =2)
print(collapsed.shape)

plt.set_cmap('gray')
plt.imshow(collapsed,cmap='gray')
plt.axis('off')
```

```python
# Evenly sampled image
even = collapsed[::2,::2] # uniform sampling

print(even.shape)
plt.imshow(even)
plt.axis('off')
plt.show()
```

No problem. The image is of the same aspect ratio (pixel height/width) as the collapsed.
Lets see what happenes if we try to show an unevenly sampled image

```python
# Evenly sampled image
uneven = collapsed[::4,::2] # uniform sampling

print(uneven.shape)
plt.imshow(uneven)
plt.axis('off')
```

The dimensions look distorded! To fix this, we need to tell the plt.imshow that the current image has been subsampled acorss the two dimensions. This can be done by passing the height/width sampling ratio as the *aspect* argument. I this case since we skipped 4 pixels in the y-dimension (height) and 2 in the x-direction (width), our aspect ratio is 2. 

```python
plt.imshow(uneven,aspect=2.0)
plt.axis('off')
```

Another way to fix the aspect of the image is to specify original image dimnesion in the optional keyword argument *extent*

```python
plt.imshow(uneven,extent=(0,480,0,305))
plt.axis('off')
```
