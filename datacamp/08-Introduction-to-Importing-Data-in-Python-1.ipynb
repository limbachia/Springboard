# **Introduction to relational database**

### **Querying data from a database**

import the connecting engine from sqlalchemy  
> from sqlalchemy import create_engine  

create an engine to connect to the database  
> engine = create_engine('sqlite:///Northwind.sqlite')

connect to the database  
> con = engine.connect()

Query the database. This gets all rows and all columns of table "Orders" from the database
> rs = con.execute("SELECT * FROM Orders")

the following turns the rs into a dataframe
> df = pd.DataFrame(rs.fetchall()) 

df does not contain column names; to assign column names to the df
> df.columns = rs.keys()

following close connection to the database
> con.close()


# To avoid forgetting to close the connection, use context manager
> with engine.connect() as con:  
>    rs = con.execute("SELECT OrderID, OrderDate FROM Orders")  
>    df = pd.DataFrame(rs.fetchmany(size=5))  
>    df.columns = rs.keys()  

in the execute method, specific column names can be specified instead of *.
even fetchmany(size=5) can be used to fetch only first 5 rows of the dataframe instead of # fetchall

### **Querying data from a database directly using pandas**

> pd.read_sql_query("SELECT * FROM Orders", engine)

### **Some use options that can be used with SELECT FROM command**

Selects all columns from table, selects only those records in which col1 is greater than or equal to 3, and orders the records by col2
> "SELECT * FROM table WHERE col1 >= 3 ORDER BY col2"

Selects all columns from table1 and table2 while doing an inner join on col1 of both the tables
> "SELECT * FROM table1 INNER JOIN table2 on table1.col1 = table2.col1"
